```python
import pandas as pd

url = "https://raw.githubusercontent.com/hanjes15/dataset_superstore_simple-analysis/refs/heads/main/Dataset_superstore_simple.csv"
df = pd.read_csv(url)

# Check first few rows
df.head()
```

```python
df.describe()
```

```python
df.columns
```

```python
# Convert to string first
df['sales'] = df['sales'].astype(str)
df['profit'] = df['profit'].astype(str)

# Keep only numbers and decimal points
df['sales'] = df['sales'].str.replace('[^0-9.-]', '', regex=True)
df['profit'] = df['profit'].str.replace('[^0-9.-]', '', regex=True)

# Convert to numeric, invalid entries become NaN
df['sales'] = pd.to_numeric(df['sales'], errors='coerce')
df['profit'] = pd.to_numeric(df['profit'], errors='coerce')

# Drop rows with missing sales or profit
df = df.dropna(subset=['sales', 'profit'])
```

```python
#Total Sales & Profit
print("Total Sales:", df['sales'].sum())
print("Total Profit:", df['profit'].sum())
```

```python
#op 5 Sub-Categories by Sales
top_sub = df.groupby('sub_category')['sales'].sum().sort_values(ascending=False).head()
print(top_sub)
```

```python
#Sales by Segment
segment_sales = df.groupby('segment')['sales'].sum()
print(segment_sales)
```

```python
#Sales by Segment
import matplotlib.pyplot as plt

# Group by segment and sum sales
segment_sales = df.groupby('segment')['sales'].sum()

# Simple bar chart
segment_sales.plot(kind='bar', title='Sales by Segment', color='skyblue')
plt.ylabel('Total Sales')
plt.show()
```

```python
#Top 5 Sub-Categories (Bar Chart
top_sub = df.groupby('sub_category')['sales'].sum().sort_values(ascending=False).head(5)

top_sub.plot(kind='bar', title='Top 5 Sub-Categories by Sales', color='orange')
plt.ylabel('Sales')
plt.show()

```

```python
segment_profit = df.groupby('segment')['profit'].sum()

segment_profit.plot(kind='bar', title='Profit by Segment', color='green')
plt.ylabel('Profit')
plt.show()
```

## Key Insights

- Consumer segment has the highest profit.
- Corporate segment is the second most profitable.
- Phones and Chairs are the two highest selling sub categories.
